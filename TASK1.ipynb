## Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading Dataset via seaborn

```python
# Load Iris dataset directly using seaborn
iris = sns.load_dataset("iris")
```

## Printing shape, column names, and the first few rows using .head().

```python
# Shape of the dataset
print("Shape:", iris.shape)

# Column names
print("Columns:", iris.columns.tolist())

# First five rows
print(iris.head())
```

## Using .info() and .describe() for summary statistics.

```python
# Info about data types and non-null values
print("\nDataset Info:")
print(iris.info())

# Descriptive statistics
print("\nDescriptive Statistics:")
print(iris.describe())
```

# Visualization

## Scatter plot

```python
sns.pairplot(iris, hue="species", height=2.5)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

## Histograms

```python
iris.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(12, 6))
for i, column in enumerate(iris.columns[:-1], 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=iris[column], color='skyblue')
    plt.title(column)
plt.suptitle("Box Plots for Outlier Detection", y=1.02)
plt.tight_layout()
plt.show()
```
